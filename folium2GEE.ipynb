```python
# import Google earth engine module
import ee

#folium
from general.myfolium import *

#import data
from general.importData import importData

#Authenticate the Google earth engine with google account
ee.Initialize()

print('Initialized')
```

```python
# Create a folium map object.
location=[28.5973518, 83.54495724]
mapObject = foliumInitialize(location,6,600)
my_map = mapObject.Initialize()

print('folium map Initialized')
```

```python
startyear = 2019
endyear = 2019

startDate = ee.Date.fromYMD(startyear,1,1)
endDate = ee.Date.fromYMD(endyear,12,31)

#gee assets to get the study area
studyArea = ee.FeatureCollection('users/bikesbade/bankey/banke')

print("getting images")
s2 = importData(studyArea,startDate,endDate)
```

```python
# Add custom basemaps
#initialize the base map class
basemaps = addbasemaps(my_map)

#for google map
basemaps.googleMap()

#for google Satellite
basemaps.GoogleSatellite()

#for google Terrain
basemaps.GoogleTerrain()

#for google Hybrid
basemaps.GoogleSatelliteHybrid()

print('basemap added')
```

```python
#Add map to layer to Folium 
my_map.add_ee_layer(s2, {'min':0,'max':0.6,'bands':"red,green,nir"}, "Sentinel")
print('Map Added')
```

```python
#initalize the plugin
tools = pluginsTools(my_map)

#layer control 
tools.addLayerControl()

#Add addFullscreen 
tools.addFullscreen()

#Add LocateControl 
tools.addLocateControl()

#Add MousePosition 
tools.addMousePosition()

#Add MeasureControl 
tools.addMeasureControl()

#Add DrawTool 
tools.addDrawTool()


print('tools added')
```

```python
startdate = "2022-07-01"

enddate = "2022-07-19"

geometry = ee.Geometry.Polygon([[[11.080417771330582, 76.86697184464953],[11.081091615882327, 77.04275308325141],[10.955055789956816, 76.98232828248202]]])

#gee assets to get the study area

S2 = ee.ImageCollection('COPERNICUS/S2').filterBounds(geometry).filterDate('2022-01-20', '2022-01-30')
# S2 = ee.ImageCollection('COPERNICUS/S2').filterDate(startdate,enddate);

print("Getting images") 

```

```python
S2.getInfo()
```

```python

#Function to calculate and add an NDVI band
def addNDVI(image):
    return image.addBands(image.normalizedDifference(['B8', 'B4']));


#Add NDVI band to image collection
S2 = S2.map(addNDVI);
SentinalNdvi = S2.select('nd')
```

```python
# Set visualization parameters.
visParams = { 'min': 0.0,
  'max': 8000.0,
  'palette': [
    'FFFFFF', 'CE7E45', 'DF923D', 'F1B555', 'FCD163', '99B718', '74A901',
    '66A000', '529400', '3E8601', '207401', '056201', '004C00', '023B01',
    '012E01', '011D01', '011301'
  ],}


# Add the data to the map object.
my_map.add_ee_layer(SentinalNdvi, visParams , 'Sential NDVI')

# Display the map.
display(my_map)
```
